# Correlation Scoring Function

```python
import gc
rescale_Y_to_mean0_std1 = True

def correlation_score(y_true, y_pred):
    """Scores the predictions according to the competition rules. 
    
    It is assumed that the predictions are not constant.
    
    Returns the average of each sample's Pearson correlation coefficient"""
    
    # Input should be matrices - does not make sense for vectors 
    if  len(y_pred.shape)< 2: return -10 # Some result to inform for incorrect input
    if  y_pred.shape[1] < 2: return -10 # Some result to inform for incorrect input

    y2 = y_pred.copy()
    y2 -= y2.mean(axis=1).reshape(-1, 1);    y2 /= y2.std(axis=1).reshape(-1, 1)    
    if rescale_Y_to_mean0_std1:
        y1 = y_true # Already rescaled 
    else:
        y1 = y_true.copy(); 
        y1 -= y1.mean(axis=1).reshape(-1, 1);    y1 /= y1.std(axis=1).reshape(-1, 1) 
        
    c = (y1*y2).mean().mean()# Correlation for rescaled matrices is just matrix product and average 
    
    c = (y1*y2).mean().mean()# Correlation for rescaled matrices is just matrix product and average 
    
    # Memory control:
    if not rescale_Y_to_mean0_std1:
        del y1
    del y2
    gc.collect()
    
    return c

    # Slow way:
    
    if type(y_true) == pd.DataFrame: y_true = y_true.values
    if type(y_pred) == pd.DataFrame: y_pred = y_pred.values
    corrsum = 0
    for i in range(len(y_true)):
        corrsum += np.corrcoef(y_true[i], y_pred[i])[1, 0]
    return corrsum / len(y_true)
```

# Correlation scores before blend

```python
%%time
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

df_stat = pd.DataFrame()
IX = 0
for i,key_loc in enumerate(list( dict_df.keys())  ):
    key_loc = list_names[i]
    df = dict_df[key_loc]
    print(key_loc, df.shape)
    #display(df.head(2))
    s = correlation_score(Y_true,df.values)
    df_stat.loc[IX,'Model'] = key_loc
    df_stat.loc[IX,'Corr'] = s
    df_stat.loc[IX,'r2'] = r2_score(Y_true,df.values)
    df_stat.loc[IX,'MSE'] = mean_squared_error(Y_true,df.values)
    
    m = mask_holdout; postfix = ' HO'
    df_stat.loc[IX,'Corr'+ postfix] = correlation_score(Y_true[m],df[m].values)
    df_stat.loc[IX,'r2'+ postfix] = r2_score(Y_true[m],df[m].values)
    df_stat.loc[IX,'MSE'+ postfix] = mean_squared_error(Y_true[m],df[m].values)
    m = mask_main;  postfix = ' Main'
    df_stat.loc[IX,'Corr'+postfix] = correlation_score(Y_true[m],df[m].values)
    df_stat.loc[IX,'r2'+postfix] = r2_score(Y_true[m],df[m].values)
    df_stat.loc[IX,'MSE'+postfix] = mean_squared_error(Y_true[m],df[m].values)
    
    IX += 1
    
display(df_stat)
```
